```python
def podstavka(st):
    list_pod=['Подставка для', 'Подставка под', 'подставка для', 'подставка под', 'Юбка под', 'Мешок для утилизации']
    list_pod_1=['Подставка держатель для елки', 'Шкатулка', 'Полянка под ёлку', 'Подставка пластиковая для', 'Подставка елочная']
    list_pod_2=['Набор', 'букет', 'енок']
    for i in (list_pod+list_pod_1+list_pod_2):
        if i in st:
            return True
    return False
```

```python
def podsvetka(st):
    list_no=['Нет', 'нет', 'Не', 'не', 'Отсу', 'отсу', 'Без', 'без', 'ожно испо']
    list_yes=['есть', 'Есть', 'C','с','да', 'Да', 'цветная', 'Гирлянда', 'гирлянда', 'лампочки']
    for i in (list_no):
        if i in st:
            return 0
    for i in (list_yes):
        if i in st:
            return 1
    return 0
```

```python
def find_cm(st):
    try:
        if 'см' in str(st):
            for i in list(range(500,10,-1)):
                if str(i) in st:
                    print(i , st)
                    return i
                if str(i) in st.replace(',', '.'):
                    print(i , st)
                    return i
            
        if 'сантиметров' in str(st):
            for i in list(range(500,10,-1)):
                if str(i) in st:
                    print(i , st)
                    return i

        return st
    except:
        return st
```

```python
def find_m(st):
    try:
        if 'м' in str(st):
            for i in [x/100 for x in range(500,10,-1)]:
                if str(i) in st:
                    print(i , st)
                    return int(100*i)
                if str(i) in st.replace(',', '.'):
                    print(i , st)
                    return int(100*i)
            
        return st
    except:
        return st
```

```python
def find_mm(st):
    try:
        if 'мм' in str(st):
            for i in list(range(10000,20,-10)):
                if str(i) in st:
                    print(i)
                    return int(i/10)
        
        return st
    except:
        return st
```
