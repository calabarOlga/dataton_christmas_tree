### ЧИСТИМ ДАННЫЕ И ВЫДЕЛЯЕМ ДОПОЛНИТЕЛЬНЫЕ ПРИЗНАКИ

В ячейке "Количество раз купили" вычленяем значение 

---





```python
def number_of_times_bought(st):
    try:
        if int(st.split(' ')[-3])>0:
            num=int(st.split(' ')[-3])*1000+int(st.split(' ')[-2])
            return num
    except:
        try:
            if int(st.split(' ')[-2])>0:
                num=int(st.split(' ')[-2])
                return num
        except:
            if st=='Error':
                return 'Error'
```

```python
df_itog['Количество раз купили']=df_itog['Количество раз купили'].apply(number_of_times_bought)
```

```python
df_itog['Количество раз купили'].unique()
```

Посчитаем предварительно количество категорий по краткому описанию из 
ячейки 'Дополнительная информация'


---



```python
words={}
n_error=0
for i in range(8001):
    try:
        for j in df_itog['Описание'][i][2:-2].split('Дополнительная информация')[1].split('\\n')[1::2]:
            
            if words.get(j, 0)==0:
                words[j]=1
            else:
                words[j]=words.get(j, 0)+1
    except:
        n_error+=1
```

Количество ошибок полученных при парсинге данных со страниц товара

---



```python
n_error
```

```python
words
```

Посчитаем количество фитчей в дополнительном описании

---



```python
words_dop={x: 0 for x in list(words.keys())}
```

```python
for i in range(len(df_itog['Описание'])):
    try:
        for word in df_itog['Описание'][i][2:-2].split('Дополнительная информация')[2].split('\\n'):
            k=0
            for fitch in list(words_dop.keys()):

                if fitch in word:
                    k+=1
                    words_dop[fitch]=words_dop.get(fitch, 0)+1
                 
            if k==0:
                words_dop[word]=words_dop.get(word, 0)+1
                               
    except:
        continue
```

```python
set(words_dop)-set(words)
```

Критерием для итогово списка признаков, которые мы оставим, будет не менее 1000

---





```python
feature_list=[x for x in list(words_dop.keys()) if words_dop[x]>1000 and words_dop[x]<10000]
```

```python
feature_list
```

```python
len(feature_list)
```

```python
def add_dict(description):
    feature_list=['Конструкция елки',
 'Подсветка',
 'Страна производства',
 'Высота упаковки',
 'Длина упаковки',
 'Особенности елки',
 'Тип елки',
 'Высота елки',
 'Ширина упаковки',
 'Вес с упаковкой (кг)',
 'Назначение подарка',
 'Материал елки',
 'Повод',
 'Количество ветвей',
 'Комплектация']
    
    feature_list_cell=[None]*15

    try:
        #длина фитчей в предварительном описании
        pred_len=len(description[2:-2].split('Дополнительная информация')[1].split('\\n')[2::2]) 
        
        #длина фитчей в полном описаниие        
        poln_len=len(description[2:-2].split('Дополнительная информация')[2].split('\\n')[1:])

        #добавляем в список все фитчи в предварительном описании
        for i in range(len(feature_list)):
            for j in range(pred_len):

                if feature_list[i]==description[2:-2].split('Дополнительная информация')[1].split('\\n')[1::2][j]:
                    feature_list_cell[i]=(description[2:-2].split('Дополнительная информация')[1].split('\\n')[2::2][j])

        #проверяем остались ли фитчи в дополнительном описании и добавляем их
        if pred_len < poln_len:             
            for i in range(len(feature_list)):
                hm=pred_len-poln_len
                for j in range(hm, 0, 1):
                    st=feature_list[i]

                    if st in description[2:-2].split('Дополнительная информация')[2].split('\\n')[1:][j]:
                        feature_list_cell[i]=description[2:-2].split('Дополнительная информация')[2].split('\\n')[1:][j].split(st+' ')[1]
        return {x:y for x,y in zip(feature_list, feature_list_cell)}  
        
    except:
        return {x:y for x,y in zip(feature_list, ['Error']*15)}
```

```python
df_itog['Список с значениями']=df_itog['Описание'].apply(add_dict)
```

#### Удаляем ячейки со значениями Error

```python
df_itog=df_itog[(df_itog['Количество раз купили']!='Error')&(df_itog['Описание']!='Error')]
```
